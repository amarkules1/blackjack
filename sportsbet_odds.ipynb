```python
# given american line odds, calculate implied percentage
def american_to_probability(odds):
    if odds > 0:
        return 100 / (odds + 100)
    else:
        return abs(odds) / (100 - odds)
```

```python
def american_to_probability_average(odds, opp_odds):
    odds_pct = american_to_probability(odds)
    opp_odds_pct = american_to_probability(opp_odds)
    return (odds_pct) / (odds_pct + opp_odds_pct)
```

```python
def profit_from_win(odds, bet=100):
    if odds > 0:
        return (odds / 100) * bet
    else:
        return 100 * (bet / abs(odds))
```

```python
# what proportion of your total should go to each side to minimize total loss?
# is it just bet size * probability of outcome? Yes
probability = american_to_probability_average(-137, 120)
total_bet = 100
bet_1 = total_bet * probability
bet_2 = total_bet * (1 - probability)
profit_if_1_wins = profit_from_win(-137, bet_1) + bet_1
profit_if_2_wins = profit_from_win(120, bet_2) + bet_2
print(profit_if_1_wins, profit_if_2_wins)
```
