```python
import pandas as pd
```

```python
# given american line odds, calculate implied percentage
def american_to_percentage(odds):
    if odds > 0:
        return 100 / (odds + 100)
    else:
        return abs(odds) / (100 - odds)
```

```python
american_to_percentage(-200)
```

```python
def american_to_percentage_average(odds, opp_odds):
    odds_pct = american_to_percentage(odds)
    opp_odds_pct = american_to_percentage(opp_odds)
    return (odds_pct) / (odds_pct + opp_odds_pct)
```

```python
american_to_percentage_average(-137, 120)
```
